### Download a sample and view - start with Electronics

- The requests.get() method downloads the gzipped JSONL files from the specified URLs.
- The gzip module decompresses the downloaded content.
- The pd.read_json() function reads the JSONL data into a pandas DataFrame with lines=True to specify that the data is in line-delimited format.
- Finally, the first few rows of each DataFrame are printed using head().

Note: This code assumes that the JSONL files are accessible at the provided URLs and that they are in the expected format. Adjustments may be necessary based on the actual structure of the data. These links worked at the time of writing this code

````python
import pandas as pd
import requests
import gzip
import os
import io
from tabulate import tabulate  # Import tabulate for pretty printing

# Function to download and load a gzipped JSONL file into a DataFrame with sampling
def load_jsonl_to_dataframe(url, sample_size=None):
    response = requests.get(url)
    if response.status_code == 200:
        with gzip.GzipFile(fileobj=io.BytesIO(response.content)) as f:
            df = pd.read_json(f, lines=True)
            if sample_size is not None:
                return df.sample(n=sample_size, random_state=1)
            return df
    else:
        raise Exception(f"Failed to download data: {response.status_code}")

# Function to save DataFrame to a specified directory
def save_dataframe_to_csv(df, file_path):
    df.to_csv(file_path, index=False)
    print(f"Data saved to {file_path}")

# URLs for the Digital Music category
review_url = "https://datarepo.eng.ucsd.edu/mcauley_group/data/amazon_2023/raw/review_categories/Digital_Music.jsonl.gz"
meta_url = "https://datarepo.eng.ucsd.edu/mcauley_group/data/amazon_2023/raw/meta_categories/meta_Digital_Music.jsonl.gz"

# Ask user for the sample size
sample_size = 100000

# Ask user for the directory to save the datasets
directory = '../data/'

# Create the directory if it doesn't exist
if not os.path.exists(directory):
    os.makedirs(directory)

# Load both the review and meta data into DataFrames with the specified sample size
digital_music_reviews_df = load_jsonl_to_dataframe(review_url, sample_size=sample_size)
digital_music_meta_df = load_jsonl_to_dataframe(meta_url, sample_size=sample_size)

# Save the DataFrames to CSV files in the specified directory
reviews_file_path = os.path.join(directory, "Digital_Music_Reviews.csv")
meta_file_path = os.path.join(directory, "Digital_Music_Meta.csv")

save_dataframe_to_csv(digital_music_reviews_df, reviews_file_path)
save_dataframe_to_csv(digital_music_meta_df, meta_file_path)

# Display the first few rows of each DataFrame in a beautiful way
print("\nDigital Music Reviews DataFrame:")
print(tabulate(digital_music_reviews_df.head(), headers='keys', tablefmt='pretty'))

print("\nDigital Music Meta DataFrame:")
print(tabulate(digital_music_meta_df.head(), headers='keys', tablefmt='pretty'))
````

````python
#### Download All Data

import os
import pandas as pd
import requests
import gzip

# Base URL for the data repository
BASE_URL = "https://datarepo.eng.ucsd.edu/mcauley_group/data/amazon_2023/raw"

# List of available categories
CATEGORIES = [
    "Electronics",
    "Books",
    "Clothing",
    "Health",
    "Home",
    "Toys",
    "Video",
    "Appliances",
    "Automotive",
    "Computers",
    "Grocery",
    "Kindle_Store",
    "Musical_Instruments",
    "Pet_Supplies",
    "Sports",
    "Tools",
    "Baby",
    "Cell_Phones",
    "Fashion",
    "Office_Products",
    "Industrial",
    "Software",
    "Jewelry",
    "Luggage",
    "Beauty",
    "Watches",
    "Arts_Crafts_Sewing",
    "Video_Games",
    "Health_and_Personal_Care",
    "Furniture"
]

def download_data(category, save_dir, review=True, meta=True):
    # Create the save directory if it doesn't exist
    os.makedirs(save_dir, exist_ok=True)

    # Construct the URLs based on the category
    review_url = f"{BASE_URL}/review_categories/{category}.jsonl.gz"
    meta_url = f"{BASE_URL}/meta_categories/meta_{category}.jsonl.gz"
    
    # Download review data if selected
    if review:
        try:
            response = requests.get(review_url)
            response.raise_for_status()  # Raise an error for bad responses
            with gzip.GzipFile(fileobj=response.content) as f:
                reviews_df = pd.read_json(f, lines=True)
            reviews_df.to_csv(os.path.join(save_dir, f"{category}_reviews.csv"), index=False)
            print(f"Downloaded reviews for {category} and saved to {save_dir}/{category}_reviews.csv")
        except Exception as e:
            print(f"Failed to download reviews for {category}: {e}")

    # Download meta data if selected
    if meta:
        try:
            response = requests.get(meta_url)
            response.raise_for_status()  # Raise an error for bad responses
            with gzip.GzipFile(fileobj=response.content) as f:
                meta_df = pd.read_json(f, lines=True)
            meta_df.to_csv(os.path.join(save_dir, f"{category}_meta.csv"), index=False)
            print(f"Downloaded meta for {category} and saved to {save_dir}/{category}_meta.csv")
        except Exception as e:
            print(f"Failed to download meta for {category}: {e}")

def main(save_dir):
    print("Available categories:")
    for idx, category in enumerate(CATEGORIES):
        print(f"{idx + 1}. {category}")

    category_input = input("Enter the category number (or type 'all' for all categories): ").strip()
    if category_input.lower() == 'all':
        selected_categories = CATEGORIES
    else:
        try:
            selected_categories = [CATEGORIES[int(category_input) - 1]]
        except (IndexError, ValueError):
            print("Invalid input. Exiting.")
            return

    review_input = input("Do you want to download reviews? (yes/no): ").strip().lower() == 'yes'
    meta_input = input("Do you want to download meta data? (yes/no): ").strip().lower() == 'yes'

    save_dir = input("Enter the directory to save the downloaded files (default: 'downloads'): ").strip() or "downloads"

    for category in selected_categories:
        download_data(category, save_dir, review=review_input, meta=meta_input, )

# Uncomment the following line to run the main function in a notebook
save_dir = '../data/'
# main()
````
